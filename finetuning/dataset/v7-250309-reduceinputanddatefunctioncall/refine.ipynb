```python
import numpy as np
import json
from pathlib import Path
import shutil
from collections import defaultdict
```

```python
base_dir = Path("./")
histogram = defaultdict(lambda: defaultdict(lambda: 0))
for scenario_dir in base_dir.iterdir():
    if not scenario_dir.is_dir() or "scenario" not in scenario_dir.name:
        continue
    # print(scenario_dir.as_posix())
    scenario_name = scenario_dir.name
    for data_file in scenario_dir.glob("onlyq_*.json"):
        
        with open(data_file, "r", encoding="utf-8") as f:
            datas = json.load(f)
            refined_datas = []
            for data in datas:
                instructions = data["Response"]["Instructions"]
                refined_instructions = []
                for instruction in instructions:
                    if instruction["type"] == "o":
                        # returns = instruction["returns"]
                        # script = instruction["script"].split(";")
                        # remove blank lines
                        # script = [line for line in script if line.strip()]
                        # instruction = {
                        #     "type": "o",
                        #     # "returns": returns,
                        #     "script": script,
                        # }

                        del instruction["returns"]
                    refined_instructions.append(instruction)
                
                data["Response"]["Instructions"] = refined_instructions
                refined_datas.append(data)
                
            with open(data_file, "w", encoding="utf-8") as f:
                json.dump(refined_datas, f, ensure_ascii=False, indent=4)
```
